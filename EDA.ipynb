### 3.1.3.2 Imputing Valuesfor Null Values (With Mean or Median???) :

```python
print("Filling the column FLOORSMAX_AVG with Mean and Median and checking how it affects Mean, Median and Mode.\n")

#Imputing the value using Mean
print("1. Filling the column FLOORSMAX_AVG with Mean and checking deviation in Mean, Median, Mode")
print("Mean : "+str(app_data_cp['FLOORSMAX_AVG'].mean())+", Median : "+str(app_data_cp['FLOORSMAX_AVG'].median())+", Mode : "+str(app_data_cp['FLOORSMAX_AVG'].mode()))
print((app_data_cp['FLOORSMAX_AVG'].fillna(value=app_data_cp['FLOORSMAX_AVG'].mean() , inplace=False)).mean(),
     (app_data_cp['FLOORSMAX_AVG'].fillna(value=app_data_cp['FLOORSMAX_AVG'].mean() , inplace=False)).median(),
     (app_data_cp['FLOORSMAX_AVG'].fillna(value=app_data_cp['FLOORSMAX_AVG'].mean() , inplace=False)).mode())
print("\n\n")

#Imputing the value using Median
print("2. Filling the column FLOORSMAX_AVG with Median and checking deviation in Mean, Median, Mode")
print("Mean : "+str(app_data_cp['FLOORSMAX_AVG'].mean())+", Median : "+str(app_data_cp['FLOORSMAX_AVG'].median())+", Mode : "+str(app_data_cp['FLOORSMAX_AVG'].mode()))
print((app_data_cp['FLOORSMAX_AVG'].fillna(value=app_data_cp['FLOORSMAX_AVG'].median() , inplace=False)).mean(),
     (app_data_cp['FLOORSMAX_AVG'].fillna(value=app_data_cp['FLOORSMAX_AVG'].median() , inplace=False)).median(),
     (app_data_cp['FLOORSMAX_AVG'].fillna(value=app_data_cp['FLOORSMAX_AVG'].median() , inplace=False)).mode())
print("\n\n")
```

As we can clearly see from the above analysis that imputing the column 'FLOORSMAX_AVG' with the mode value does not affect the Mean, Median and Mode, it is safer to fill it with the Mode Value. This is done in the next code block.

```python
print("1. Number of Null items in that column before imputing "+str(app_data_cp['FLOORSMAX_AVG'].isnull().sum()))
print("2. Null percentage for that column before imputing "+str(100*(app_data_cp['FLOORSMAX_AVG'].isnull().sum())/(len(app_data_cp['FLOORSMAX_AVG'].index))))
app_data_cp['FLOORSMAX_AVG'].fillna(value=app_data_cp['FLOORSMAX_AVG'].median() , inplace=True)
print("\n\n1. Number of Null items in that column after imputing "+str(app_data_cp['FLOORSMAX_AVG'].isnull().sum()))
print("2. Null percentage for that column after imputing "+str(100*(app_data_cp['FLOORSMAX_AVG'].isnull().sum())/(len(app_data_cp['FLOORSMAX_AVG'].index))))
```

#### 3.1.5.2 Outlier identification using IQR Method :

```python
for i in (app_data_cp2.columns).tolist():
    if((app_data_cp2[i].dtype=='float64' or app_data_cp2[i].dtype=='int64') and (len(app_data_cp2[i].value_counts())>10)):
        perc=np.abs(app_data_cp2[i].quantile([0.25,0.75]))
        low = perc[0.25] - 1.5*(perc[0.75]-perc[0.25])
        high = perc[0.75] + 1.5*(perc[0.75]-perc[0.25])
        print("Column Name: "+i+"\n 25th Quantile: "+str(perc[0.25])+", 75th Quantile: "+str(perc[0.75])+"\n Lower Limit: "+str(low)+", Higher Limit: "+str(high))
        print("Number of rows/items/elements below the bottom outlier : "+str(len(app_data_cp2[i][(app_data_cp2[i]<low)].index)))
        print("Number of rows/items/elements above the top outlier : "+str(len(app_data_cp2[i][(app_data_cp2[i]>high)].index))+"\n")
```

#### 3.1.5.3 Handling The Identified Outlier using Trimming Technique :
###### Using this technique so that we get the most kind of people who apprach for loan

```python
#Handling the Outlier by trimming method for the column CNT_CHILDREN(Count Of Children) and AMT_REQ_CREDIT_BUREAU_QRT
import math

app_data_outlier_trimming=app_data_cp2.copy()



print("---------------------------------- TRIMMING CNT_CHILDREN BASED ON IQR BASED OUTLIER ANALYIS ----------------------------------")

perc=abs(app_data_outlier_trimming['CNT_CHILDREN'].quantile([0.25,0.75]))

#Calculating the boundary values based on IQR method
low_cnt_children = math.floor(perc[0.25] - 1.5*(perc[0.75]-perc[0.25]))
high_cnt_children = math.ceil(perc[0.75] + 1.5*(perc[0.75]-perc[0.25]))
print("1. The lower boundary value is "+str(low_cnt_children)+" and the upper boundary value is "+str(high_cnt_children))

#Getting the list of row index to drop them based on the boundary values calculated
inde = app_data_outlier_trimming['CNT_CHILDREN'][(app_data_outlier_trimming['CNT_CHILDREN']<low_cnt_children) | (app_data_outlier_trimming['CNT_CHILDREN']>high_cnt_children)].index
print("\n2. The number of rows that are part of the outlier is {0} and its percentage in the dataframe is {1}.".format(len(inde),(len(inde)*100/len(app_data_outlier_trimming.index))))
app_data_outlier_trimming.drop(inde, inplace=True)
print("\n3. The distinct values and their counts for the column CNT_CHILDREN is")
print(app_data_outlier_trimming.CNT_CHILDREN.value_counts())



print("\n\n\n------------------------ TRIMMING AMT_REQ_CREDIT_BUREAU_QRT BASED ON IQR BASED OUTLIER ANALYIS ------------------------")

perc=abs(app_data_outlier_trimming['AMT_REQ_CREDIT_BUREAU_QRT'].quantile([0.1,0.9]))

#Calculating the boundary values based on IQR method
low_AMT_REQ_CREDIT_BUREAU_QRT = math.floor(perc[0.1] - 1.5*(perc[0.9]-perc[0.1]))
high_AMT_REQ_CREDIT_BUREAU_QRT = math.ceil(perc[0.9] + 1.5*(perc[0.9]-perc[0.1]))
print("1. The lower boundary value is "+str(low_AMT_REQ_CREDIT_BUREAU_QRT)+" and the upper boundary value is "+str(high_AMT_REQ_CREDIT_BUREAU_QRT))

#Getting the list of row index to drop them based on the boundary values calculated
inde2 = app_data_outlier_trimming['AMT_REQ_CREDIT_BUREAU_QRT'][(app_data_outlier_trimming['AMT_REQ_CREDIT_BUREAU_QRT']<low_AMT_REQ_CREDIT_BUREAU_QRT) | (app_data_outlier_trimming['AMT_REQ_CREDIT_BUREAU_QRT']>high_AMT_REQ_CREDIT_BUREAU_QRT)].index
print("\n2. The number of rows that are part of the outlier is {0} and its percentage in the dataframe is {1}.".format(len(inde2),(len(inde2)*100/len(app_data_outlier_trimming.index))))
app_data_outlier_trimming.drop(inde2, inplace=True)
print("\n3. The distinct values and their counts for the column AMT_REQ_CREDIT_BUREAU_QRT is")
print(app_data_outlier_trimming.AMT_REQ_CREDIT_BUREAU_QRT.value_counts())



print("\n\n\n---------------------------------- TRIMMING AMT_INCOME_TOTAL BASED ON IQR BASED OUTLIER ANALYIS ----------------------------------")

perc=abs(app_data_outlier_trimming['AMT_INCOME_TOTAL'].quantile([0.01,0.99]))

#Calculating the boundary values based on IQR method
low_AMT_INCOME_TOTAL = math.floor(perc[0.01] - 1.5*(perc[0.99]-perc[0.01]))
high_AMT_INCOME_TOTAL = math.ceil(perc[0.99] + 1.5*(perc[0.99]-perc[0.01]))
print("1. The lower boundary value is "+str(low_AMT_INCOME_TOTAL)+" and the upper boundary value is "+str(high_AMT_INCOME_TOTAL))

#Getting the list of row index to drop them based on the boundary values calculated
inde3 = app_data_outlier_trimming['AMT_INCOME_TOTAL'][(app_data_outlier_trimming['AMT_INCOME_TOTAL']<low_AMT_INCOME_TOTAL) | (app_data_outlier_trimming['AMT_INCOME_TOTAL']>high_AMT_INCOME_TOTAL)].index
print("\n2. The number of rows that are part of the outlier is {0} and its percentage in the dataframe is {1}.".format(len(inde3),(len(inde3)*100/len(app_data_outlier_trimming.index))))
app_data_outlier_trimming.drop(inde3, inplace=True)
print("\n3. The Maximum value after trimming the AMT_INCOME_TOTAL column is "+str(app_data_outlier_trimming.AMT_INCOME_TOTAL.max()))


print("\n\n\n---------------------------------- TRIMMING YEARS_EMPLOYED BASED ON IQR BASED OUTLIER ANALYIS ----------------------------------")

perc=abs(app_data_outlier_trimming['YEARS_EMPLOYED'].quantile([0.25,0.75]))

#Calculating the boundary values based on IQR method
low_YEARS_EMPLOYED = math.floor(perc[0.25] - 1.5*(perc[0.75]-perc[0.25]))
high_YEARS_EMPLOYED = math.ceil(perc[0.75] + 1.5*(perc[0.75]-perc[0.25]))
print("1. The lower boundary value is "+str(low_YEARS_EMPLOYED)+" and the upper boundary value is "+str(high_YEARS_EMPLOYED))

#Getting the list of row index to drop them based on the boundary values calculated
inde4 = app_data_outlier_trimming['YEARS_EMPLOYED'][(app_data_outlier_trimming['YEARS_EMPLOYED']<low_YEARS_EMPLOYED) | (app_data_outlier_trimming['YEARS_EMPLOYED']>high_YEARS_EMPLOYED)].index
print("\n2. The number of rows that are part of the outlier is {0} and its percentage in the dataframe is {1}.".format(len(inde4),(len(inde4)*100/len(app_data_outlier_trimming.index))))
app_data_outlier_trimming.drop(inde4, inplace=True)

print("\n3. The distinct values and their counts for the column YEARS_EMPLOYED is")
print(app_data_outlier_trimming.YEARS_EMPLOYED.value_counts())



print("\n\n")
print("The number of rows before trimming "+str(len(app_data_cp2.index)))
print("The number of rows after trimming "+str(len(app_data_outlier_trimming.index)))
print("The number of rows that were trimmed "+str(len(app_data_cp2.index) - len(app_data_outlier_trimming.index)))
```
